```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/kanee/Downloads/Data_Analyst_Assignment_Dataset.csv",header=0)
```

```python
def get_bounce_info(bounce_string):
    if not isinstance(bounce_string, str):
        return 0, 0  # Handle non-string values gracefully
    
    bounce_count = sum(b == 'B' or b == 'L' for b in bounce_string)
    last_bounce_month = int(bounce_string[-2]) if bounce_string[-1] in ['B', 'L'] and bounce_string[-2].isdigit() else 0
    
    return bounce_count, last_bounce_month
```

```python
df['Bounce Count'], df['Last Bounce Month'] = zip(*df['Bounce String'].apply(get_bounce_info))
```

```python
bounce_counts, last_bounce_months = zip(*df['Bounce String'].apply(get_bounce_info))


```

```python

df['Bounce Count'] = bounce_counts
df['Last Bounce Month'] = last_bounce_months
```

```python
df['Last Bounce Month'] = df['Last Bounce Month'].replace('N', np.nan)
```

```python
df['Last Bounce Month'] = df['Last Bounce Month'].fillna(0).astype(int)
```

```python
df['Months Since Last Bounce'] = (df['Tenure'] - df['Bounce Count'] - df['Last Bounce Month'])
```

```python
def assign_risk_label(row):
    if row['Bounce Count'] == 0:
        return 'Unknown Risk'
    elif row['Months Since Last Bounce'] >= 6:
        return 'Low Risk'
    elif row['Bounce Count'] <= 1 and row['Months Since Last Bounce'] >= 2:
        return 'Medium Risk'
    else:
        return 'High Risk'
```

```python
df['Risk Label'] = df.apply(assign_risk_label, axis=1)
```

```python
def categorize_tenure(row):
    if row['Tenure'] <= 3:
        return 'Early Tenure'
    elif row['Tenure'] >= (df['Tenure'].max() - 3):
        return 'Late Tenure'
    else:
        return 'Mid Tenure'
```

```python
df['Tenure Category'] = df.apply(categorize_tenure, axis=1)
```

```python
print(df.head())
```

```python
tenure_categories = df['Tenure Category'].value_counts()
plt.pie(tenure_categories.values, labels=tenure_categories.index, autopct='%1.1f%%')
plt.axis('equal')
plt.xlabel('Tenure Category')
plt.ylabel('Count')
plt.title('Distribution of Customers by Tenure Category')
plt.show()
```

```python
df['Ticket Size'] = pd.qcut(df['Amount Pending'], q=3, labels=['Low', 'Medium', 'High'])
```

```python
def assign_resource(row):
    if row['Bounce String'][0] == 'S':
        if row['Amount Pending'] <= 1000:
            return 'Digital'
        elif row['Amount Pending'] <= 5000:
            return 'Voice Bot'
        else:
            return 'Human Calling'
    elif row['Bounce String'][0] == 'B':
        if row['Amount Pending'] <= 1000:
            return 'Voice Bot'
        elif row['Amount Pending'] <= 5000:
            return 'Human Calling'
        else:
            return 'Human Calling'
    else:
        return 'Human Calling'
```

```python
df['Resource'] = df.apply(assign_resource, axis=1)
```
